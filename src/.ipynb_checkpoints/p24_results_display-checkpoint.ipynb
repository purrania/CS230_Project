```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_accuracy_from_csv(file_name, title="Training and Validation Accuracy"):
    # Load the CSV file
    data = pd.read_csv(file_name)
    epochs = range(1, len(data) + 1)  # Assuming each row corresponds to one epoch
    accuracy = data['accuracy']
    val_accuracy = data['val_accuracy']
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, accuracy, label='Training Accuracy', marker='o')
    plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(title)  # Use the custom title
    plt.legend()
    plt.grid(True)

    plt.show()

plot_accuracy_from_csv('cifar-100_base_history.csv', title='Cifar-100 base') 
```

```python
plot_accuracy_from_csv('cifar-100_activation_history.csv', title='Cifar-100 Evolution') 
```

```python
plot_accuracy_from_csv('cifar-100_reset_activation_history.csv', title='Cifar-100 Reset') 
```

```python
plot_accuracy_from_csv("cifar-10_base_history.csv")
```
