```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_accuracy_from_csv(file_name, title="Training and Validation Accuracy"):
    # Load the CSV file
    data = pd.read_csv(file_name)
    epochs = range(1, len(data) + 1)  # Assuming each row corresponds to one epoch
    accuracy = data['accuracy']
    val_accuracy = data['val_accuracy']
    plt.figure(figsize=(10, 6))
    plt.plot(epochs, accuracy, label='Training Accuracy', marker='o')
    plt.plot(epochs, val_accuracy, label='Validation Accuracy', marker='o')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(title)  # Use the custom title
    plt.legend()
    plt.grid(True)

    plt.show()

#plot_accuracy_from_csv('cifar-100_base_history.csv', title='Cifar-100 base') 
```

```python
plot_accuracy_from_csv('cifar-100 Baseline Model.csv', title='Cifar-100 Baseline Model') 
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
prefixes = ["CIFAR-10", "CIFAR-100", "Fashion MNIST"]
suffixes = ["Baseline Model.csv", "Growing Model.csv", "Mature Model.csv"]
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
directory = './'
for prefix in prefixes:
    filenames = []
    for suffix in suffixes:
        filenames.append(prefix + " " + suffix)
    plt.figure(figsize=(10, 6))
    for idx, file in enumerate(filenames):
        filepath =  filepath = os.path.join(directory, file)
        data = pd.read_csv(filepath)
        if 'accuracy' in data.columns and 'val_accuracy' in data.columns:
            color = colors[idx % len(colors)]  # Cycle through colors
            plt.plot(data['accuracy'], label=f"{file} - Accuracy", linestyle='-', color=color)
            plt.plot(data['val_accuracy'], label=f"{file} - Val Accuracy", linestyle='--', color=color)
        else:
            print(f"Missing required columns in file '{file}'. Skipping.")
    plt.title(f"Accuracy Trends for {prefix}")
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)

    # Save the plot as a file
    plot_filename = f"{prefix}{suffix.replace('.csv', '')}_accuracy_plot.png"
    plt.savefig(plot_filename)
    plt.show()
```
